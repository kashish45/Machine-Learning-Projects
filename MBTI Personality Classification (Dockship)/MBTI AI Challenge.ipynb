```python
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv('TRAIN.csv')
test_data=pd.read_csv('TEST.csv')
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
train_data.fillna(" ",inplace=True)
```

```python
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
train_data.drop(['idx'],axis=1,inplace=True)
test_data.drop(['idx'],axis=1,inplace=True)
```

```python
data=train_data.iloc[:,1:]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)

#Removing column names
list1=[i for i in range(len(data.columns))]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head(5)
```

```python
for index in new_index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:]))
```

```python
headlines[2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#Implementing Bag Of words
countVector=CountVectorizer(ngram_range=(5,5))
traindata=countVector.fit_transform(headlines)
```

```python
    rf=RandomForestClassifier(n_estimators=200,criterion='entropy')
rf.fit(traindata,train_data['type'])
```

```python
rf.score(traindata,train_data['type'])
```

```python
test_transform=[]
for row in range(0,len(test_data.index)):
    test_transform.append(' '.join(str(x) for x in test_data.iloc[row,0:]))
testdataset=countVector.transform(test_transform)
predictions=rf.predict(testdataset)
```
